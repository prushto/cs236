```python
from torchvision.transforms import functional as F
def preprocess_image(image):
    image = torch.tensor(image).unsqueeze(0)
    #print(image.shape)
    image = image / 255.0
    return F.center_crop(image, (256, 256))
```

```python
baseline_images = [
    #transform(Image.open('wellness.png')),
    preprocess_image(transform(Image.open('train_1.png').convert("RGB"))),
    preprocess_image(transform(Image.open('train_2.png').convert("RGB"))),
    #transform(Image.open('fashion_brand.png')),
    preprocess_image(transform(Image.open('train_3.png').convert("RGB")))
]
final_tensor_baseline = torch.stack(baseline_images)
final_baseline_images = torch.cat([image for image in final_tensor_baseline])

#print(final_baseline_images.shape)

baseline1_images = [
    #transform(Image.open('wellness.png')),
    preprocess_image(transform(Image.open('baseline1_train_1.png'))),
    preprocess_image(transform(Image.open('baseline1_train_2.png'))),
    #transform(Image.open('fashion_brand.png')),
    preprocess_image(transform(Image.open('baseline1_train_3.png')))
]
final_tensor_baseline1 = torch.stack(baseline1_images)
final_baseline1_images = torch.cat([image for image in final_tensor_baseline1])
#final_baseline_images = final_baseline_images.permute(0, 3, 1, 2)
#print(final_baseline2_images.shape)

fid = FrechetInceptionDistance(normalize=True)
fid.update(final_baseline_images, real=True)
fid.update(final_baseline1_images, real=False)

print(f"FID: {float(fid.compute())}")
```
