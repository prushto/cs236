```python
from torchmetrics.functional.multimodal import clip_score
from functools import partial

clip_score_fn = partial(clip_score, model_name_or_path="openai/clip-vit-base-patch16")

def calculate_clip_score(images, prompts):
    images_int = (images * 255).astype("uint8")
    clip_score = clip_score_fn(torch.from_numpy(images_int).permute(0, 3, 1, 2), prompts).detach()
    return round(float(clip_score), 4)
```

FID

```python
from torchvision.transforms import functional as F
def preprocess_image(image):
    image = torch.tensor(image).unsqueeze(0)
    #print(image.shape)
    image = image / 255.0
    return F.center_crop(image, (256, 256))
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
])
```

```python
baseline_images = [
    #transform(Image.open('wellness.png')),
    preprocess_image(transform(Image.open('train_1_new.png').convert("RGB"))),
    preprocess_image(transform(Image.open('train_2_new.png').convert("RGB"))),
    #transform(Image.open('fashion_brand.png')),
    preprocess_image(transform(Image.open('train_3_new.png').convert("RGB")))
]
final_tensor_baseline = torch.stack(baseline_images)
final_baseline_images = torch.cat([image for image in final_tensor_baseline])

#print(final_baseline_images.shape)

baseline1_images = [
    #transform(Image.open('wellness.png')),
    preprocess_image(transform(Image.open('myra-astro-dreambooth-2.png'))),
    preprocess_image(transform(Image.open('myra-cyberpunk-dreambooth-2.png'))),
    #transform(Image.open('fashion_brand.png')),
    preprocess_image(transform(Image.open('myra-jedi-dreambooth-2.png')))
]
final_tensor_baseline1 = torch.stack(baseline1_images)
final_baseline1_images = torch.cat([image for image in final_tensor_baseline1])
#final_baseline_images = final_baseline_images.permute(0, 3, 1, 2)
#print(final_baseline2_images.shape)

fid = FrechetInceptionDistance(normalize=True)
fid.update(final_baseline_images, real=True)
fid.update(final_baseline1_images, real=False)

print(f"FID: {float(fid.compute())}")
```
